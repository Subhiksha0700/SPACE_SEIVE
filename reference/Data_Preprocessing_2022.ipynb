```python
train = data_train.copy()
```

```python
data_test['class'].value_counts()
```

```python
new_train = train[train['class']=='debris']


list = data_train['class'].value_counts().index.tolist()
list.remove('debris')

class_data = pd.DataFrame()
for class_name in list:
    data = train[train['class'] == class_name].head(600)
    class_data = pd.concat([class_data, data])

class_data = class_data.sample(frac=1, random_state=42)
class_data.reset_index(drop=True, inplace=True)

Filtered_train = pd.concat([new_train, class_data])
Filtered_train = Filtered_train.sample(frac=1, random_state=42)
Filtered_train.reset_index(drop=True, inplace=True)
Filtered_train

```

```python
test = data_test.copy()
new_test = test[test['class']=='debris']

list_test = test['class'].value_counts().index.tolist()
list_test.remove('debris')

class_data_test = pd.DataFrame()
for class_name in list:
    data1 = test[test['class'] == class_name].head(200)
    class_data_test = pd.concat([class_data_test, data1])

class_data_test = class_data_test.sample(frac=1, random_state=42)
class_data_test.reset_index(drop=True, inplace=True)

Filtered_val = pd.concat([new_test, class_data_test])
Filtered_val = Filtered_val.sample(frac=1, random_state=42)
Filtered_val.reset_index(drop=True, inplace=True)
Filtered_val
```

```python
Filtered_train['filename'] = Filtered_train['filename'].str[:-4]
Filtered_train['filename'] = Filtered_train['filename'] + '.jpg'
Filtered_train

Filtered_val['filename'] = Filtered_val['filename'].str[:-4]
Filtered_val['filename'] = Filtered_val['filename'] + '_2022.jpg'
Filtered_val
```

```python
category_to_zero = 'debris'
Filtered_train['class'] = Filtered_train['class'].apply(lambda x: 0 if x == category_to_zero else 1)
Filtered_val['class'] = Filtered_val['class'].apply(lambda x: 0 if x == category_to_zero else 1)
```

```python
import os
import shutil

filenames_train_df = pd.DataFrame()

filenames_train_df['filename'] = Filtered_train['filename']

folder_2022_train = "/Users/nithish/Documents/Academics/Spring 24/AML/project/data/spark-2022-stream-1/train"
master_train_folder = "/Users/nithish/Documents/Academics/Spring 24/AML/project/data/master_data/train"

# Iterate through the list of filenames and transfer corresponding images
for filename in filenames_train_df['filename']:
    
    finalName = filename
    filename = filename[:-9]
    filename = filename +'.jpg'
    source_path = os.path.join(folder_2022_train, filename)
    destination_path = os.path.join(master_train_folder, finalName)
    if os.path.exists(source_path):
        shutil.copy(source_path, destination_path)

print("Images transferred.")
```

```python
filenames_test_df = pd.DataFrame()

filenames_test_df['filename'] = Filtered_val['filename']

folder_2022_val = "/Users/nithish/Documents/Academics/Spring 24/AML/project/data/spark-2022-stream-1/val"
master_val_folder = "/Users/nithish/Documents/Academics/Spring 24/AML/project/data/master_data/val"

# Iterate through the list of filenames and transfer corresponding images
for filename in filenames_test_df['filename']:
    finalName = filename
    filename = filename[:-9]
    filename = filename +'.jpg'
    source_path = os.path.join(folder_2022_val, filename)
    destination_path = os.path.join(master_val_folder, finalName)
    if os.path.exists(source_path):
        shutil.copy(source_path, destination_path)

print("Images transferred.")
```
